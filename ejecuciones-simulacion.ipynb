```python
from dataclasses import dataclass

from simulacion import meses, simulacion


@dataclass
class SimParameter:
    days: int
    price: int


params = [SimParameter(days=1, price=10), SimParameter(days=1, price=15), SimParameter(days=1, price=20),
          SimParameter(days=1, price=25), SimParameter(days=1, price=30),
          SimParameter(days=2, price=10), SimParameter(days=2, price=15), SimParameter(days=2, price=20),
          SimParameter(days=2, price=25), SimParameter(days=2, price=30),
          SimParameter(days=3, price=10), SimParameter(days=3, price=15), SimParameter(days=3, price=20),
          SimParameter(days=3, price=25), SimParameter(days=3, price=30),
          SimParameter(days=4, price=10), SimParameter(days=4, price=15), SimParameter(days=4, price=20),
          SimParameter(days=4, price=25), SimParameter(days=4, price=30),
          SimParameter(days=7, price=10), SimParameter(days=7, price=15), SimParameter(days=7, price=20),
          SimParameter(days=7, price=25), SimParameter(days=7, price=30),
          SimParameter(days=15, price=10), SimParameter(days=15, price=15), SimParameter(days=15, price=20),
          SimParameter(days=15, price=25), SimParameter(days=15, price=30),
          SimParameter(days=30, price=10), SimParameter(days=30, price=15), SimParameter(days=30, price=20),
          SimParameter(days=30, price=25), SimParameter(days=30, price=30),
          ]
```

```python
resultados = [simulacion(tiempoFinal=meses(240), intervaloEntreSubidas=60 * 24 * p.days, precioSuscripcion=p.price) for
              p
              in params]
```

```python

@dataclass
class Results:
    gananciasAnuales: float
    engagement: float
    porcentajeDePropinas: float
    costoDeProduccionAnual: float
    precio: int
    dias: int
    label: str
```

```python
results = [
    Results(gananciasAnuales=r.gananciasAnuales,
            engagement=r.engagement,
            porcentajeDePropinas=r.porcentajeDePropinas,
            costoDeProduccionAnual=r.costoDeProduccionAnual,
            precio=p.price,
            dias=p.days,
            label=f"({p.days}, {p.price})"
            ) for r, p in zip(resultados, params)]
```

```python
df['tips'] = df['gananciasAnuales'] * df['porcentajeDePropinas'] / 100
```

```python

import matplotlib.pyplot as plt

# Group by 'dias' and get the row with the max 'gananciasAnuales' for each group
max_rows = df.iloc[df.groupby('dias')['gananciasAnuales'].idxmax().values.flatten()].drop_duplicates()
# Create bar plot
bar_width = 0.4

bars_ganancias = plt.bar(max_rows['label'], max_rows['gananciasAnuales'], width=bar_width, align='center',
                         label='Maxima Ganancias Anuales [USD]')
bars_tips = plt.bar(max_rows['label'], max_rows['tips'], width=bar_width, align='edge', label='Propinas Anuales [USD]')

plt.xlabel('Intervalo [dias], Precio Subscripcion [USD]')
plt.legend()
plt.show()
```

```python
# Group by 'dias' and get the row with the max 'amount' and 'percentage' for each group
max_rows = df.iloc[df.groupby('dias')['costoDeProduccionAnual'].idxmax().values.flatten()].drop_duplicates()
# Create bar plot
bar_width = 0.4

plt.bar(max_rows['label'], max_rows['costoDeProduccionAnual'], width=bar_width, align='center',
        label='Maximo Costo Anual [USD]')
plt.xlabel('Intervalo [dias], Precio Subscripcion [USD]')
plt.legend()
plt.show()
```

```python
# Group by 'dias' and get the row with the max 'amount' and 'percentage' for each group
max_rows = df.iloc[df.groupby('dias')['engagement'].idxmax().values.flatten()].drop_duplicates()
# Create bar plot
bar_width = 0.4

plt.bar(max_rows['dias'], max_rows['engagement'], width=bar_width, align='center',
        label='Maximo Engagement [%]')
plt.xlabel('Intervalo [dias]')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(27, 14))
ax = fig.add_subplot(111)
plt.xlabel('Intervalo [dias]; Precio [USD]')
plt.bar(df['label'], df['gananciasAnuales'], label='Ganancias Anuales [USD]')
plt.bar(df['label'], df['tips'], label='Propinas Anuales [USD]')
plt.bar(df['label'], df['costoDeProduccionAnual'], label='Costo Anual [USD]')
plt.legend()
plt.show()
```

```python
engagements = [simulacion(tiempoFinal=meses(24), intervaloEntreSubidas=60 * 24 * i, precioSuscripcion=20)
               for i in range(1, 31)]
```

```python
engagements_results = [
    Results(gananciasAnuales=r.gananciasAnuales,
            engagement=r.engagement,
            porcentajeDePropinas=r.porcentajeDePropinas,
            costoDeProduccionAnual=r.costoDeProduccionAnual,
            precio=20,
            dias=d,
            label=f"{d}"
            ) for r, d in zip(engagements, range(1, 31))]
```

```python
df_engagements = pandas.DataFrame(engagements_results)
df_engagements
```

```python
df_engagements.sort_values('dias', ascending=True, inplace=True)
cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']
plt.figure(figsize=(15, 8))
plt.xlabel('Intervalo [dias]')
plt.bar(df_engagements['label'], df_engagements['engagement'], color=cycle[2], label='Engagement')
plt.axhline(55, color=cycle[1], linestyle='--', label='Min. Top Creator')
plt.legend()
plt.show()
```
